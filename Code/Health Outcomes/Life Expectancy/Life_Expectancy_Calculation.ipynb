# Calculating new life table with adjusted mortality rate

```python
lifetable = pd.read_csv(r'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/lifetable_unified.csv')

lifetable.rename(columns={'region':'iso_code'}, inplace=True)

lifetable
```

```python
mortalityrate_adjusted = pd.read_csv('/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Adjusted_Mortality_Rate/mortalityrate_adjusted.csv')

mortalityrate_adjusted
```

```python
lifetable_ex = lifetable[['scen', 'Time', 'sex', 'iso_code', 'agest', 'sx', 'lx', 'Lx', 'ax', 'dx','Tx', 'ex', 'mx']].merge\
    (mortalityrate_adjusted[['scen', 'Time', 'sex', 'iso_code', 'agest', 'mx_new_AVG','mx_new_LOW','mx_new_HIGH', 'varvaldesc','Scenario']]\
        , on=['scen', 'Time', 'sex','iso_code','agest'],how='outer').dropna(subset=['mx_new_AVG']).reset_index(drop=True)

lifetable_ex
```

```python
def process_life_table(df):

    df1 = df[['scen', 'Time', 'sex', 'iso_code', 'agest', 'sx', 'lx', 'Lx',
              'dx', 'Tx', 'ex', 'mx', 'varvaldesc', 'Scenario']].copy()

    df = df[df['mx'] != 0][['Scenario', 'Time', 'sex', 'iso_code', 'agest', 'ax', 
                             'mx_new_AVG', 'mx_new_LOW', 'mx_new_HIGH']].copy()


    def calculate_new_metrics(df, col_name_suffix, mx_col):

        df[f'mx.new_{col_name_suffix}'] = df[mx_col].astype(float)


        df[f'qx.new_{col_name_suffix}'] = 5 * df[f'mx.new_{col_name_suffix}'] / \
                                          (1 + (5 - df['ax']) * df[f'mx.new_{col_name_suffix}'])
        df.loc[df.index[-1], f'qx.new_{col_name_suffix}'] = 1  # Last age interval qx = 1


        df[f'px.new_{col_name_suffix}'] = 1 - df[f'qx.new_{col_name_suffix}']


        l0 = 100000
        df[f'lx.new_{col_name_suffix}'] = [float(l0)] + [float('nan')] * (len(df) - 1)


        for i in range(1, len(df)):
            df.iloc[i, df.columns.get_loc(f'lx.new_{col_name_suffix}')] = (
                df.iloc[i - 1, df.columns.get_loc(f'lx.new_{col_name_suffix}')] * 
                df.iloc[i - 1, df.columns.get_loc(f'px.new_{col_name_suffix}')]
            )


        df[f'dx.new_{col_name_suffix}'] = df[f'lx.new_{col_name_suffix}'] * df[f'qx.new_{col_name_suffix}']


        df[f'Lx.new_{col_name_suffix}'] = 5 * df[f'lx.new_{col_name_suffix}'] + \
                                          (df['ax'] - 5) * df[f'dx.new_{col_name_suffix}']
        df.loc[df.index[-1], f'Lx.new_{col_name_suffix}'] = (
            df.loc[df.index[-1], f'lx.new_{col_name_suffix}'] / df.loc[df.index[-1], f'mx.new_{col_name_suffix}']
        )  


        df[f'sx.new_{col_name_suffix}'] = 0.0


        for i in range(0, len(df) - 1):
            df.iloc[i, df.columns.get_loc(f'sx.new_{col_name_suffix}')] = (
                df.iloc[i + 1, df.columns.get_loc(f'Lx.new_{col_name_suffix}')] / 
                df.iloc[i, df.columns.get_loc(f'Lx.new_{col_name_suffix}')]
            )
        df.loc[df.index[-1], f'sx.new_{col_name_suffix}'] = 0  


        df[f'Tx.new_{col_name_suffix}'] = df[f'Lx.new_{col_name_suffix}'][::-1].cumsum()


        df[f'ex.new_{col_name_suffix}'] = df[f'Tx.new_{col_name_suffix}'] / df[f'lx.new_{col_name_suffix}']

    calculate_new_metrics(df, 'AVG', 'mx_new_AVG')
    calculate_new_metrics(df, 'LOW', 'mx_new_LOW')
    calculate_new_metrics(df, 'HIGH', 'mx_new_HIGH')

    df = df1.merge(df, on=['Time', 'sex', 'iso_code', 'agest', 'Scenario'], how='left').fillna(0)

    return df
```

```python
group_columns = ['Scenario', 'sex', 'iso_code', 'Time']
lifetable_ex_adjusted = (
    lifetable_ex.groupby(group_columns, group_keys=False)  
    .apply(process_life_table)                   
)


lifetable_ex_adjusted.reset_index(drop=True, inplace=True) 

for i in ['AVG','LOW','HIGH']:

    lifetable_ex_adjusted[f'ex_diff_{i}'] = lifetable_ex_adjusted[f'ex.new_{i}'] - lifetable_ex_adjusted['ex']
    

lifetable_ex_adjusted
```
