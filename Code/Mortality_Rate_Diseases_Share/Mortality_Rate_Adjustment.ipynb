# Unifiy the 2020 life table using WPP2022

```python
wpp_2022_lt_m = pd.read_excel(
    r'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Data/UN/life table/WPP2022_MORT_F07_2_ABRIDGED_LIFE_TABLE_MALE.xlsx',
    sheet_name='Estimates', skiprows=16, index_col=None)

wpp_2022_lt_f = pd.read_excel(
    r'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Data/UN/life table/WPP2022_MORT_F07_3_ABRIDGED_LIFE_TABLE_FEMALE.xlsx',
    sheet_name='Estimates', skiprows=16, index_col=None)

wpp_2022_lt_m_2020 = wpp_2022_lt_m[wpp_2022_lt_m['Year'] == 2020]
wpp_2022_lt_m_2020 = wpp_2022_lt_m_2020.dropna(subset=['ISO3 Alpha-code']).drop(
    columns=['Index', 'Variant', 'Notes', 'Location code', 'ISO2 Alpha-code', 'SDMX code**', 'Type', 'Parent code']
).reset_index(drop=True)

wpp_2022_lt_m_2020.rename(columns={
    'Region, subregion, country or area *': 'country name', 'ISO3 Alpha-code': 'region', 'Year': 'Time',
    'Age (x)': 'agest', 'Central death rate m(x,n)': 'mx', 'Probability of dying q(x,n)': 'qx',
    'Probability of surviving p(x,n)': 'px', 'Number of survivors l(x)': 'lx', 'Number of deaths d(x,n)': 'dx',
    'Number of person-years lived L(x,n)': 'Lx', 'Survival ratio S(x,n)': 'sx', 'Person-years lived T(x)': 'Tx',
    'Expectation of life e(x)': 'ex', 'Average number of years lived a(x,n)': 'ax'
}, inplace=True)

wpp_2022_lt_f_2020 = wpp_2022_lt_f[wpp_2022_lt_f['Year'] == 2020]
wpp_2022_lt_f_2020 = wpp_2022_lt_f_2020.dropna(subset=['ISO3 Alpha-code']).drop(
    columns=['Index', 'Variant', 'Notes', 'Location code', 'ISO2 Alpha-code', 'SDMX code**', 'Type', 'Parent code']
).reset_index(drop=True)

wpp_2022_lt_f_2020.rename(columns={
    'Region, subregion, country or area *': 'country name', 'ISO3 Alpha-code': 'region', 'Year': 'Time',
    'Age (x)': 'agest', 'Central death rate m(x,n)': 'mx', 'Probability of dying q(x,n)': 'qx',
    'Probability of surviving p(x,n)': 'px', 'Number of survivors l(x)': 'lx', 'Number of deaths d(x,n)': 'dx',
    'Number of person-years lived L(x,n)': 'Lx', 'Survival ratio S(x,n)': 'sx', 'Person-years lived T(x)': 'Tx',
    'Expectation of life e(x)': 'ex', 'Average number of years lived a(x,n)': 'ax'
}, inplace=True)

def merge_age_groups(group):
    group['agest'] = group['agest'].astype(int)
    age_0, age_1, age_5 = group[group['agest'] == 0], group[group['agest'] == 1], group[group['agest'] == 5]
    if age_0.empty or age_1.empty or age_5.empty:
        return group
    
    lx_0 = age_0['lx'].values[0]
    dx_05 = age_0['dx'].values[0] + age_1['dx'].values[0]
    Lx_05 = age_0['Lx'].values[0] + age_1['Lx'].values[0]
    mx_05 = dx_05 / Lx_05
    qx_05 = 1 - (age_0['px'].values[0] * age_1['px'].values[0])
    px_05 = 1 - qx_05
    sx_05 = age_5['Lx'].values[0] / Lx_05
    Tx_05 = Lx_05 + age_5['Tx'].values[0]
    ex_05 = Tx_05 / lx_0
    ax_05 = ((age_0['ax'].values[0] * age_0['dx'].values[0]) + ((age_1['ax'].values[0] + 1) * age_1['dx'].values[0])) / dx_05

    new_row = pd.DataFrame([{
        'country name': age_0['country name'].values[0],
        'region': age_0['region'].values[0],
        'Time': age_0['Time'].values[0],
        'agest': '0-5',
        'Age interval (n)': 5,
        'mx': mx_05, 'qx': qx_05, 'px': px_05, 'lx': lx_0, 'dx': dx_05,
        'Lx': Lx_05, 'sx': sx_05, 'Tx': Tx_05, 'ex': ex_05, 'ax': ax_05
    }])

    group = group[~group['agest'].isin(['0', '1'])].copy()
    group = pd.concat([group, new_row], ignore_index=True)
    return group

df_m = wpp_2022_lt_m_2020.copy()
df_m_updated = df_m.groupby('region', group_keys=False).apply(merge_age_groups)
df_m_updated['sex'] = 'm'
df_m_updated = df_m_updated[(df_m_updated['agest'] != 0) & (df_m_updated['agest'] != 1)].drop(columns=['country name', 'Age interval (n)'])
df_m_updated['agest'] = df_m_updated['agest'].replace({'0-5': 0}).astype('int64')
df_m_updated = df_m_updated.sort_values(by=['region', 'agest']).reset_index(drop=True)

scenarios = ['SSP1', 'SSP2', 'SSP3']
df_m_updated = pd.concat([df_m_updated.drop(columns=['px', 'qx']).assign(scen=scen) for scen in scenarios], ignore_index=True)

df_f = wpp_2022_lt_f_2020.copy()
df_f_updated = df_f.groupby('region', group_keys=False).apply(merge_age_groups)
df_f_updated['sex'] = 'f'
df_f_updated = df_f_updated[(df_f_updated['agest'] != 0) & (df_f_updated['agest'] != 1)].drop(columns=['country name', 'Age interval (n)'])
df_f_updated['agest'] = df_f_updated['agest'].replace({'0-5': 0}).astype('int64')
df_f_updated = df_f_updated.sort_values(by=['region', 'agest']).reset_index(drop=True)
df_f_updated = pd.concat([df_f_updated.drop(columns=['px', 'qx']).assign(scen=scen) for scen in scenarios], ignore_index=True)
```

# Disaggregate mortality rates to disease-specific level

```python
disease_paths = {
    'COP': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/COP_data.csv',
    'IHD': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/IHD_data.csv',
    'LRI': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/LRI_data.csv',
    'LUN': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/LUN_data.csv',
    'STR': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/STR_data.csv',
    'TDM': '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/Foresight/share_of_diseases/TDM_data.csv'
}

disease_data = {}
for disease, path in disease_paths.items():
    disease_data[disease] = pd.read_csv(path)

    disease_data[disease] = disease_data[disease][disease_data[disease]['age'] != 'All']

    disease_data[disease]['age'] = disease_data[disease]['age'].replace({'95+': 95})

    disease_data[disease]['age'] = disease_data[disease]['age'].astype('int')

for ssps in ['SSP1','SSP2','SSP3']:
    dataframe_name = f'mortalityrate_{ssps}'
    globals()[dataframe_name] = mortalityrate[mortalityrate['SSPs'] == ssps].reset_index(drop=True)
    
    globals()[dataframe_name]['age'] = globals()[dataframe_name]['agest'].replace({100: 95})

for ssps in ['SSP1', 'SSP2', 'SSP3']:
    for disease, disease_df in disease_data.items():

        mortalityrate_name = f'mortalityrate_{ssps}'

        merged_df = globals()[mortalityrate_name].merge(disease_df, left_on=['year', 'sex', 'location', 'age'], right_on=['year', 'sex', 'location', 'age'], how='outer')
        
        merged_df.drop(columns=['age'], inplace=True)
        merged_df.rename(columns={'agest': 'age',
                                  'location':'iso_code',
                                  'year':'Year',
                                  'mean': f'{disease}_AVG',
                                'lower': f'{disease}_LOW',
                                'upper': f'{disease}_HIGH'}, inplace=True)
        merged_df.dropna(subset=['cause','SSPs','mx'],axis=0,inplace=True)
        merged_df.sort_values(by=['iso_code', 'sex', 'age'], inplace=True)
        merged_df['cause']= f'{disease}'

        merged_df.reset_index(drop=True)
        
        output_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Mortality Rates_Diseases Share/{ssps}_mortalityrate_{disease}.csv'
        merged_df.to_csv(output_path, index=False)
        
        globals()[f'{ssps}_mortalityrate_{disease}'] = merged_df
```

# Adjusting mortality rate with PM2.5 feedbacks effect

```python
replacement_dict = {100: 95}

diseases = ['IHD', 'STR', 'LRI', 'LUN', 'TDM', 'COP']

scenarios = ['ssp119', 'ssp126', 'ssp245', 'ssp370']

def process_disease(scenario, disease, scenario_mortality_change):
    
    if scenario == 'ssp119' or scenario == 'ssp126':
        mortality_data_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Mortality Rates_Diseases Share/SSP1_mortalityrate_{disease}.csv'
    elif scenario == 'ssp245':
        mortality_data_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Mortality Rates_Diseases Share/SSP2_mortalityrate_{disease}.csv'
    elif scenario == 'ssp370':
        mortality_data_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Mortality Rates_Diseases Share/SSP3_mortalityrate_{disease}.csv'

    mortality_df = pd.read_csv(mortality_data_path)
    
    rr_data_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/GBD 2021/PM25_RR_AF_reference_ssp245/PM25_RR_AF_reference_ssp245_{scenario}_{disease}.csv'
    
    rr_df = pd.read_csv(rr_data_path)

    rr_df = rr_df[['iso_code', 'Year', 'RR_AVG_reference', 'RR_LOW_reference','RR_HIGH_reference']]
    
    merged_df = mortality_df.merge(rr_df, on=['iso_code', 'Year'], how='outer')

    merged_df.rename(columns={
        'RR_AVG_reference': f'{disease}_RR_AVG_reference',
        'RR_LOW_reference': f'{disease}_RR_LOW_reference',
        'RR_HIGH_reference': f'{disease}_RR_HIGH_reference'

    }, inplace=True)
    
    merged_df.dropna(subset=['SSPs', f'{disease}_RR_AVG_reference'], axis=0, inplace=True)

    if disease == 'LRI':
        pass
    else:
        merged_df.loc[merged_df['age'] < 25, merged_df.columns.str.contains('RR_')] = 1
        merged_df.loc[merged_df['age'] < 25, merged_df.columns.str.contains('AF_')] = 0
    
    merged_df.drop(columns=['cause'], inplace=True)
    
    for i in ['AVG', 'LOW', 'HIGH']:
        merged_df[f'{disease}_mx_change_{i}'] = merged_df['mx'] * merged_df[f'{disease}_{i}'] * merged_df[f'{disease}_RR_{i}_reference']
        merged_df[f'{disease}_mx_unchange_{i}'] = merged_df['mx'] * merged_df[f'{disease}_{i}']

    if scenario_mortality_change is None:
        scenario_mortality_change = merged_df
    else:
        common_columns = ['SSPs', 'Year', 'sex', 'iso_code', 'age', 'mx']
        scenario_mortality_change = scenario_mortality_change.merge(merged_df, on=common_columns, how='outer')

    return scenario_mortality_change

all_scenarios_df = []

for scenario in scenarios:

    scenario_mortality_change = None

    for disease in diseases:
        scenario_mortality_change = process_disease(scenario, disease, scenario_mortality_change)
        if scenario_mortality_change is None:
            raise ValueError(f"Processing disease {disease} resulted in None.")

    for i in ['AVG', 'LOW', 'HIGH']:

        scenario_mortality_change[f'mx_affected_{i}'] = scenario_mortality_change[[f'{disease}_mx_change_{i}' for disease in diseases]].sum(axis=1)
        scenario_mortality_change[f'mx_affected_original_{i}'] = scenario_mortality_change[[f'{disease}_mx_unchange_{i}' for disease in diseases]].sum(axis=1)


        total_disease_fraction = scenario_mortality_change[[f'{disease}_{i}' for disease in diseases]].sum(axis=1)
        scenario_mortality_change[f'mx_unaffected_{i}'] = scenario_mortality_change['mx'] * (1 - total_disease_fraction)
        

        scenario_mortality_change[f'mx_new_{i}'] = scenario_mortality_change[f'mx_affected_{i}'] + scenario_mortality_change[f'mx_unaffected_{i}']

    columns_to_keep = ['SSPs', 'Year', 'sex', 'iso_code', 'age', 'mx']
    for i in ['AVG', 'LOW', 'HIGH']:
        columns_to_keep.extend([f'mx_affected_{i}',f'mx_affected_original_{i}', f'mx_unaffected_{i}',  f'mx_new_{i}'])

    scenario_mortality_change = scenario_mortality_change[columns_to_keep]

    scenario_mortality_new_df = scenario_mortality_change.merge(lt_region_recode, on=['iso_code'], how='left').drop(columns=['varval']).drop_duplicates()

    scenario_mortality_new_df.rename(columns={
        'SSPs': 'scen',
        'Year': 'Time',
        'age': 'agest'
    }, inplace=True)
    
    scenario_mortality_new_df['Scenario'] = scenario
    
    all_scenarios_df.append(scenario_mortality_new_df)
    
    output_path = f'/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Adjusted_Mortality_Rate/{scenario}_mortalityrate_adjusted.csv'
    scenario_mortality_new_df.to_csv(output_path, index=False)


mortalityrate_adjusted = pd.concat(all_scenarios_df, ignore_index=True)

for i in ['AVG', 'LOW', 'HIGH']:
    mortalityrate_adjusted.loc[mortalityrate_adjusted[f'mx_new_{i}'] > 1, f'mx_new_{i}'] = 0.9999999999999999

output_path = '/Users/kediliukl/surfdrive/Documents/CML research/IIASA-YSSP/Liu et al code and dataset /Code&Data for results/Data/SSP datasets/Life Tables/Adjusted_Mortality_Rate/mortalityrate_adjusted.csv'
mortalityrate_adjusted.to_csv(output_path, index=False)

mortalityrate_adjusted
```
